**TOPIC**: DIABETES PREDICTION ON FEMALE PATIENTS 

**BACKGROUND STUDY**

Diabetes is a chronic metabolic disease characterized by high blood sugar levels. This occurs when the body either doesn't produce enough insulin or can't effectively use the insulin it produces. Insulin is a hormone that regulates blood glucose, and its dysfunction leads to a buildup of sugar in the bloodstream, potentially damaging various organs over time. Over time,uncontrolled diabetes can cause severe problems within the body’s systems, including the heart, blood vessels, eyes, kidneys, and nerves. There is no cure for diabetes.


**STAGES OF DIABETES**:
- Type 1 diabetes: It is an autoimmune disease where the immune system attacks and destroys beta cells in the pancreas, which are responsible for producing insulin.
- Type 2 diabetes : It is a chronic condition where the body either doesn't produce enough insulin or can't properly use the insulin it produces, leading to high blood sugar levels. At this stage, muscle, fat, and liver cells become resistant to insulin, requiring the pancreas to produce more insulin to maintain blood sugar levels within the normal range.

```python
def func(c):
    if c < 18.5:
        return "underweight"
    elif (c >= 18.5 and c <= 24.9):
        return "healthyweight"
    elif (c >= 25 and c <= 29.9):
        return "overweight"
    elif(c >= 30):
        return "obesity"
df["BMI"] = df["BMI"].map(func)
```

```python
#converting age into categorical variable

def func(b):
    
    if (b >= 20 and b <= 40):
        return "youth"
    elif (b >= 41 and b <= 69):
        return "adult"
    elif (b >= 70):
        return "old"
df["Age"] = df["Age"].map(func)
```

```python
#converting blood pressure into categorical variable

def func(g):
    if (g <= 60):
        return "low"
    elif (g >= 61 and g <= 79):
        return "normal"
    elif (g >= 80 and g <= 89):
        return "prohypertension"
    elif (g >= 90 and g <= 99):
        return "stage1"
    elif (g >= 100 and g <=  109):
        return "stage2"
    elif (g >= 110):
        return "crisis"
df["BloodPressure"] = df["BloodPressure"].map(func)
```

```python
#Converting outcome into categorical variable

def func(z):
    if (z == 0):
        return "diabetic"
    else:
        return "not_diabetic"
df["Outcome"] = df["Outcome"].map(func)
```
